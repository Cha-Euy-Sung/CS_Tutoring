```python
import csv
import datetime

def readCSV(filename):
    dict_list = []
    file_ = csv.DictReader(open(filename, newline='', encoding='utf-8'))
    for content in file_:
        dict_list.append(dict(content))
    return dict_list

def writeCSV(filename, list_of_dictionaries):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        field_names = list_of_dictionaries[0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=field_names)
        writer.writeheader()
        for content in list_of_dictionaries:
            writer.writerow(content)

def keepOnly(list_of_dictionaries, key, value):
    new_list_of_dictionaries = []
    for content in list_of_dictionaries:
        if content[key] == value:
            new_list_of_dictionaries.append(content)
    return new_list_of_dictionaries

def discardOnly(list_of_dictionaries, key, value):
    new_list_of_dictionaries = []
    for content in list_of_dictionaries:
        if content[key] != value:
            new_list_of_dictionaries.append(content)
    return new_list_of_dictionaries

def filterRange(list_of_dictionaries, key, low, high):
    new_list_of_dictionaries = []
    for content in list_of_dictionaries:
        if (content[key] >= low)&(content[key]<high):
            new_list_of_dictionaries.append(content)
    return new_list_of_dictionaries

def duration(date1, date2):
    diff = date2 - date1
    return diff.days
```

```python
def departments(list_of_dictionaries):
    deptList = list()
    for content in list_of_dictionaries:
        if content["SUBJECT"] not in deptList:
            deptList.append(content["SUBJECT"])
    deptList.sort()
    return deptList
```

```python
def open_year(dictionary):
    return int(dictionary["OPEN DATE"][6:10])  
```

```python
def filterYear(list_of_dictionaries, low, high):
    newList = list()
    for dictionary in list_of_dictionaries:
        year = open_year(dictionary)
        if year >= int(low) and year < int(high):
            newList.append(dictionary)
    return newList

filterYear(list_of_dictionaries, 2017, 2018)
```
